Melalui eksplor ini ditemukan pula nilai missing values dengan status order sudah terkirim. Nilai minus mungkin disebabkan oleh kesalahan input dan missing values mungkin kesalahan sistem di mana, order_delivered_carrier_date	dan order_delivered_customer_date tidak tertulis.

Untuk kasus order_delivered_carrier_date dengan status delivered yang missing atau nilai delivery_time-nya negatif (minus) kita akan isi dengan order_delivered_customer_date - waktu pengiriman rata-rata. Sedangkan order_delivered_customer_date dengan status delivered yang missing akan diisi dengan prediksi sampainya barang tersebut sesuai kolom order_estimated_delivery_date 

**inputasi data order_delivered_customer_date dengan status delivered**

```python
orders_df.loc[orders_df.order_status == 'delivered', 'order_delivered_customer_date'] = \
    orders_df.loc[orders_df.order_status == 'delivered', 'order_delivered_customer_date'].fillna(
        orders_df.loc[orders_df.order_status == 'delivered', 'order_estimated_delivery_date']
    )
```

**inputasi missing_values pada order_delivered_carrier_date (status delivered)**

```python
delivery_time_mean = round(orders_df.loc[orders_df.order_status == 'delivered', 'delivery_time'].mean())
```

```python
orders_df.loc[(orders_df.order_status == 'delivered'), 'order_delivered_carrier_date'] = \
    orders_df.loc[orders_df.order_status == 'delivered', 'order_delivered_carrier_date'].fillna(
        orders_df.loc[orders_df.order_status == 'delivered', 'order_delivered_customer_date'] - pd.Timedelta(days=delivery_time_mean)
    )
```

**inputasi order_delivered_carrier_date (status delivered) yang menghasilkan delivery_time negatif**

```python
orders_df.loc[(orders_df.order_status == 'delivered') & (orders_df.delivery_time < 0), 'order_delivered_carrier_date'] = \
     orders_df.loc[orders_df.order_status == 'delivered', 'order_delivered_customer_date'] - pd.Timedelta(days=delivery_time_mean)
```

**Menghitung ulang delivery_time**

```python
delivery_time = orders_df.order_delivered_customer_date - orders_df.order_delivered_carrier_date
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df['delivery_time'] = round(delivery_time/(60 * 60 * 24))
```

### Clustering Order Reviews based on Sentiment

Di sini kita akan mengelompokkan rating_score menjadi sentiment positive, negative, dan netral. Di mana jika score kurang dari 3 artinya negative, lebih dari 3 positive, dan score sama dengan 3 itu netral.

```python
def sentimen(x):
    if x > 3:
        return 'positive'
    elif x < 3:
        return 'negative'
    elif x == 3:
        return 'netral'
    else:
        return x
all_df['sentiment'] = all_df['review_score'].apply(lambda x: sentimen(x))
all_df.head()
```

```python
all_df.loc[all_df.order_status.isin(['delivered', 'shipped'])].sentiment.isna().sum()
```

```python
sentiment_df = all_df.loc[all_df.order_status.isin(['delivered', 'shipped'])].copy()
sentiment_df.dropna(subset=['sentiment'], inplace=True)
sentiment_df
```

Distribusi Sentimen

```python
dist_sentimen = sentiment_df.groupby(by='sentiment').review_id.nunique().reset_index()
dist_sentimen.rename(columns={'review_id': 'review_count'}, inplace=True)
dist_sentimen
```

```python
plt.figure(figsize=(10, 5))

colors_ = ["#D3D3D3", "#D3D3D3", "#72BCD4"]

sns.barplot(
    y='review_count',
    x='sentiment',
    data=dist_sentimen.sort_values(by='review_count', ascending=False),
    palette=colors_,
    hue='review_count', 
    legend=False,
)

plt.title('Sentiment Review Distribution', loc='center', fontsize=15)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```
